3.Define Machine Learning. What are the main components in Machine Learning?

  -> Machine Learning (ML) is a subset of artificial intelligence (AI) that
     focuses on creating systems capable of learning and improving from experience without being explicitly programmed

  -> Main Components of Machine Learning:

  1.Data:

   Data serves as the foundation for ML models.

  2.Features:

   Features are individual measurable properties or inputs extracted from raw data that are relevant to the task.

  3.Model:

  The ML model is the mathematical representation that processes input data and predicts outputs.

  7.Deployment:

  This step involves integrating the trained model into real-world systems for practical use, such as recommending movies or detecting fraudulent transactions.

6.How do we handle categorical variables in Machine Learning? What are the
  common techniques?


  -> 1.One-Hot Encoding:
      This creates binary columns for each category in the variable. For example, if the variable "Color" has three categories (Red, Blue, Green), One-Hot Encoding will create three binary columns: "Color_Red," "Color_Blue," and "Color_Green."

     Advantage: No assumption of ordinal relationships between categories.

     Best for nominal categorical variables.

     2.Label Encoding:
      Each category is assigned a unique integer value. For example, "Red" = 1, "Blue" = 2, "Green" = 3.

      Advantage: Simple and quick.

      Limitation: This introduces an ordinal relationship which might mislead the model for non-ordinal variables.

15.What is causation? Explain difference between correlation and causation with
   an example.

  -> Causation:-
      Causation refers to a relationship where one variable directly causes a change in another.

  -> Difference Between Correlation and Causation:
     Correlation:-
       Measures the strength and direction of a relationship between two variables.
       Example: Ice cream sales and drowning incidents are correlated (both increase during summer), but one does not cause the other.

      Causation:-
       Indicates a cause-and-effect relationship where one variable directly impacts the other.
        
        Example: Pressing the accelerator of a car causes it to speed up. Here, pressing the accelerator is the direct cause of increased speed.

25.Explain data encoding?


  -> Data encoding is the process of converting data into a format suitable for
     machine learning models, especially when working with categorical or textual data that needs to be represented numerically. Since most machine learning algorithms operate on numerical inputs, encoding is essential to enable models to understand and process non-numeric data.


  -> 1.Label Encoding:
       Converts each category into a unique integer value.
     2.One-Hot Encoding:
       Creates binary columns for each category.
     3.Ordinal Encoding:
       Similar to label encoding but explicitly accounts for the order of the categories.
     5.Target Encoding:
        Replaces categories with the mean of the target variable for each category.
